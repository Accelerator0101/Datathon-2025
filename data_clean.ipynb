```python
#identify duplicate rows
df = df.dropna(subset=['Description'])
```

upon investigating duplicated rows we can see that the `receipt_id` and `date` attributes are different.

```python
neg_count = df[df['UnitPrice'] < 0].shape[0]
neg_rows = df[df['UnitPrice'] < 0]
neg_count
neg_rows
```

```python
df = df[ ~df['Description'].str.contains('adjust bad debt', case=False) ]
df = df[df['UnitPrice'] >= 0]
```

```python
df.duplicated().any()
```

```python
duplicateRows = df[df.duplicated()]
duplicateRows
```

```python
sorted_df = df.sort_values('UnitPrice', ascending=False).reset_index(drop=True)

# 2) Export to CSV

sorted_df.head(10)
```

```python
# performing null check
df.isnull().sum()
```

```python
pattern = r'postage|amazon|manual|cruk commission|discount|bank charges|carriage|samples|Dotcomgiftshop Gift Voucher'

# 3) Keep only rows that do NOT match
clean_df = sorted_df[~sorted_df['Description']
    .str.contains(pattern, case=False, na=False)
].reset_index(drop=True)

# 4) Quick sanity check
print("Original rows:", df.shape[0])
print("Cleaned rows: ", clean_df.shape[0])

clean_df.head(10)
```

```python
df.isnull().sum()
```

```python
# 1) Mask for any letter in StockCode
mask = clean_df['StockCode'].str.contains(r'[A-Za-z]', regex=True, na=False)

# 2) Subset to StockCode + Description (and drop exact duplicates if you like)
alpha_items = clean_df.loc[mask, ['StockCode', 'Description']].drop_duplicates()

# 3) Display
print(alpha_items)
# or in Jupyter:
alpha_items.head(10)
```

```python
potage_rows = clean_df[clean_df['Description']
    .str.contains('MANUAL', case=False, na=False)
]

# 2) How many did we find?
print(f"Rows containing 'potage': {len(potage_rows)}")

# 3) Display them (or just peek)
potage_rows       # or potage_rows.head(50)
```

```python
# 1) Build a mask: StockCode with any A–Z or a–z
mask = df['StockCode'].str.contains(r'[A-Za-z]', regex=True, na=False)

# 2) Get the unique Descriptions for those rows
unique_descriptions = df.loc[mask, 'Description'].unique()
print("Unique Descriptions for alpha StockCodes:")
for desc in unique_descriptions:
    print("-", desc)

# 3) (Optional) See their frequencies
print("\nCounts:")
print(df.loc[mask, 'Description'].value_counts())
```

```python
mask = clean_df['StockCode'].str.count(r'[A-Za-z]') >= 2

# 2) Use it to extract unique descriptions
unique_desc = clean_df.loc[mask, 'Description'].unique()

# 3) Inspect the result
print(f"Found {len(unique_desc)} unique descriptions where StockCode has ≥2 letters:")
for desc in unique_desc:
    print(" -", desc)
```
